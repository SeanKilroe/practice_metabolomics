```python
import pandas as pd
import os

# 1. SETUP - We use the exact path and filename you provided
folder = r"E:\UTHSCSA files\MAC metabolomics, proteomics, lipidomics march 2026 ALL SUBS"
filename = "Copy of 2026_03_Kilroe_Human_muscle_metabolomics_v2.csv"
full_path = os.path.join(folder, filename)

# 2. LOAD DATA
df = pd.read_csv(full_path)

# 3. PICK INTERNAL STANDARD (Proline)
is_name = "Proline.IS"

# 4. EXTRACT IS DATA
# We grab the Peak Area for Proline.IS for every sample number
is_data = df[df['Metabolite'] == is_name][['LCMS Sample No.', 'Peak Area']]
is_data.columns = ['LCMS Sample No.', 'IS_Area']

# 5. MERGE & NORMALIZE
# This adds the Proline area as a column next to all other metabolites
df = df.merge(is_data, on='LCMS Sample No.')

# MATH: Raw Area / (Internal Standard Area * Protein Weight)
df['Normalized_Area'] = df['Peak Area'] / (df['IS_Area'] * df['Protein Amount (mg)'])

# 6. SAVE THE RESULTS
# This creates a new CSV in your current Jupyter folder
output_name = "Normalized_Metabolomics_Proline.csv"
df.to_csv(output_name, index=False)

print(f"Success! File saved as: {output_name}")

# --- QUICK VERIFICATION ---
# If the math worked, Proline's own Normalized_Area should be 1 / Protein Amount
print("\n--- Validation (Proline Rows) ---")
check = df[df['Metabolite'] == is_name][['Sample Name', 'Normalized_Area', 'Protein Amount (mg)']].head()
print(check)
```
